```python
# Viet for integer coefficients and а = +-1

import math

def get_viet_roots(a, b, c):
  
  divisors_list = get_c_divisors_list(a, b, c) # find all the divisors of 'с'
  pos_x_1, pos_x_2 = root_signs_def(divisors_list, a, b, c) # determine the signs of possible roots
  roots_list = roots_def(pos_x_1, pos_x_2, a, b, c) # check the possible roots on the sum '-b' and find the only pair

  return roots_list


def get_viet_roots_short(a, b, c):
  
  divisors_list = get_c_divisors_list_short(a, c) # find all the divisors of 'с'
  pos_x_1, pos_x_2 = root_signs_def(divisors_list, a, b, c) # determine the signs of possible roots
  roots_list = roots_def(pos_x_1, pos_x_2, a, b, c) # check the possible roots on the sum '-b' and find the only pair

  return roots_list


# find all the divisors of 'с'
def get_c_divisors_list(a, b, c): # x1 * x2 = c/a

  print('x1 * x2 =', int(c/a))
  divisors_list = [1, int(abs(c/a))]
  print('Find all the divisors of {}\nA priori divisors {}: {}'.format(int(abs(c/a)), int(abs(c/a)), divisors_list))
  print('No divisor of a number, other than itself, can be greater than its half\nTherefore, we will check up to {} without repetitions'.format(math.ceil(int(abs(c/a))/2)))
  for d in [d for d in range(2, math.ceil(abs(c/a)/2)+1)]: # semimax divisor
    if d not in divisors_list:
      if abs(c/a)%d==0: # will record in pairs at ones: |x2| > |x1|
        divisors_list.extend([d, int(abs(c/a)/d)])

  print('All the divisors', int(abs(c/a)), ':', divisors_list)
  print('x1 + x2 =', int(b*(-1)/a))
  return divisors_list


def get_c_divisors_list_short(a, c): # x1 * x2 = c/a

  divisors_list = [1, int(abs(c/a))]
  for d in [d for d in range(2, math.ceil(abs(c/a)/2)+1)]: # semimax divisor
    if d not in divisors_list:
      if abs(c/a)%d==0: # will record in pairs at ones
        divisors_list.extend([d, int(abs(c/a)/d)])
        
  return divisors_list


# determine the signs of possible roots
def root_signs_def(divisors_list, a, b, c):
  if c/a<0:
    print('One of the roots is negative')
  
    if b*(-1)/a<0:
      print('The negative root is modulo greater than the positive one')
      pos_x_1, pos_x_2 = divisors_list[::2], [d*(-1) for d in divisors_list[1::2]] # pairs: |x2| > |x1|
    else: # if -b>0
      print('The positive root is greater than the negative modulo')
      pos_x_1, pos_x_2 = [d*(-1) for d in divisors_list[::2]], divisors_list[1::2] # pairs: |x2| > |x1|

  else: # if c>0
    print('Both roots are positive or both are negative')
  
    if b*(-1)/a<0:
      print('Both roots are negative')
      pos_x_1, pos_x_2 = [d*(-1) for d in divisors_list[::2]], [d*(-1) for d in divisors_list[1::2]]
    else: # if -b>0
      print('Both roots are positive')
      pos_x_1, pos_x_2 = divisors_list[::2], divisors_list[1::2]

  return pos_x_1, pos_x_2


# check the possible roots on the sum '-b' and find the only pair
def roots_def(pos_x_1, pos_x_2, a, b, c):
  
  roots_list = []
  for el, r in enumerate(pos_x_1): # x1 + x2 = - b/a
    if r+pos_x_2[el]==b*(-1)/a:
      roots_list.extend([r, pos_x_2[el]])
  
  if len(roots_list)==0:
    result = 'Mistake in the coefficients'

  elif len(roots_list)>2:
    result = 'The equation has more than 2 roots - ???'

  else: # x1 * x2 = c/a
    if roots_list[0]*roots_list[1]==c/a:
      # ax^2 + bx + c = 0
      if a*(roots_list[0]**2)+b*roots_list[0]+c==0 and a*(roots_list[1]**2)+b*roots_list[1]+c==0:
        result = roots_list
    
    else:
      result = 'Could not find the roots of the equation'

  return result
```
