```python
# Decomposition of the Newton Binomial through Pascal's Triangle

import numpy as np


# Creating the Pascal Triangle of the n-th degree

def get_pascal_triangle(n): # binomial's degree

  r = int(n+1) # number of array rows
  c = r*2-1 # number of array columns
  pascal_triangle = np.zeros((r, c), dtype = int) # array of n+1 rows by n*2-1 columns filled with zeros
  pascal_triangle[0, int((c-1)/2)] = 1 # change the middle element of the first row of the array to 1
  for l in range(1, r): # for each subsequent line
    for m in range(c): # calculate the value of each element      !!!!!!!!!  is it possible to use a mask ???
      try:
        pascal_triangle[l, m] = pascal_triangle[l-1, m-1] + pascal_triangle[l-1, m+1] # the sum of the top two elements diagonally
      except: pass
  pascal_triangle[r-1, c-1] = 1 # change the last element of the last line

  return pascal_triangle


# get the decomposition of the Newton Binomial of the nth degree

# Complete function with all monomes, operands and result
def get_newton_binomial(a, operand, b, n): # (a+-b)^n

  if operand not in ['+', '-']: # check operand correctness
    return [0], [0], ['invalid operand'], [0]

  else:
    pascal_triangle = get_pascal_triangle(n) # get the Pascal Triangle of the n-th degree
    coeffs = [el for el in pascal_triangle[n] if el>0] # get the coefficients for the n-th degree

  # get all the monomes
    monoms_alg = []
    monoms_num = []
    for el, coeff in enumerate(coeffs):
      monoms_alg.append(''.join([str(coeff), '*a^', str(n), '*b^', str(el)])) # in algebraic form
      monoms_num.append(coeff*(a**n)*(b**el)) # calculated by the decomposition formula
      n -= 1

   # get the operands and calculate the result
    action_list = []
    result = 0
    result_list = []
    
    if operand == '+':
      for el, monom in enumerate(monoms_num):
        action_list.append(operand)
        result += monom
        result_list.append(result)
    
    else:
      for el, monom in enumerate(monoms_num):
        if el%2==0:
          action_list.append('+')
          result += monom
          result_list.append(result)
        else:
          action_list.append('-')
          result -= monom
          result_list.append(result)

    return monoms_alg, monoms_num, action_list, result_list


# Short function - result only
def get_newton_binomial_short(a, operand, b, n): # (a+-b)^n

  if operand not in ['+', '-']: # check operand correctness
    return 'invalid operand'

  else:
    pascal_triangle = get_pascal_triangle(n) # get the Pascal Triangle of the n-th degree
    coeffs = [el for el in pascal_triangle[n] if el>0] # get the coefficients for the n-th degree

  # get all the monomes
    monoms_num = []
    for el, coeff in enumerate(coeffs):
      monoms_num.append(coeff*(a**n)*(b**el)) # calculated by the decomposition formula
      n -= 1

   # get the operands and calculate the result
    result = 0
    if operand == '+':
      for el, monom in enumerate(monoms_num):
        result += monom
    else:
      for el, monom in enumerate(monoms_num):
        if el%2==0:
          result += monom
        else:
          result -= monom

    return result
```

```python
# Decomposition of the Newton Binomial of the n-th degree - (a+-b)^n

a = 1
b = 2
n = 10
operand = '-'

# Complete function
monoms_alg, monoms_num, action_list, result_list = get_newton_binomial(a, operand, b, n)

for el, monom in enumerate(monoms_alg):
  print(action_list[el], '->', monom, '->', monoms_num[el], '->', result_list[el])
print('')
print('result =', result_list[-1])
print('')

# Short function
result = get_newton_binomial_short(a, operand, b, n)
print(result)
```
